# ***Build a recommender system by using cosine simillarties score***.

```python
book=pd.read_csv('/content/book (1).csv', encoding='latin-1')
```

```python
book
```

```python
book00=book.iloc[1:,1:]
```

```python
book00
```

```python
book['Book.Title'].value_counts()
```

```python
a=book['User.ID'].unique()
```

```python
book1=book.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
book1
```

```python
book1.fillna(0,inplace=True)
```

```python
book1
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
simlar=1-pairwise_distances(book1,metric='cosine')
```

```python
simlar
```

```python

user_sim=pd.DataFrame(simlar)
```

```python
user_sim.index=a
user_sim.columns=a
```

```python

user_sim
```

```python

np.fill_diagonal(simlar,0)
```

```python
user_sim
```

```python
#similar book readers 
user_sim.idxmax(axis=1)
```
